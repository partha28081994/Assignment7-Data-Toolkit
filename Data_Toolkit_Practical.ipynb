# 4.How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = {
    "Variable1": [1, 2, 3, 4, 5],
    "Variable2": [5, 4, 3, 2, 1],
    "Variable3": [2, 3, 4, 5, 6]
}
df = pd.DataFrame(data)
correlation_matrix = df.corr()
print(correlation_matrix)

plt.figure(figsize=(8, 6))  
sns.heatmap(correlation_matrix, 
            annot=True,        
            cmap="coolwarm",   
            vmin=-1, vmax=1,   
            square=True,       
            linewidths=0.5)  
plt.title("Correlation Matrix Heatmap")
plt.show()
```

# 8.Create a line plot with multiple lines using Matplotlib.

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y1 = [2, 3, 5, 7, 11]
y2 = [1, 4, 6, 8, 10]
y3 = [3, 5, 8, 12, 15]

plt.figure(figsize=(8, 5))

# Plot each line
plt.plot(x, y1, label='Line 1', color='blue', linestyle='-', marker='o')
plt.plot(x, y2, label='Line 2', color='green', linestyle='--', marker='s')
plt.plot(x, y3, label='Line 3', color='red', linestyle='-.', marker='d')

plt.title("Line Plot with Multiple Lines")
plt.xlabel("X-axis Label")
plt.ylabel("Y-axis Label")
plt.legend()

plt.grid(True)
plt.show()
```

# 9.Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold.

```python
import pandas as pd

item = ["A","B","C","D","E"]
price = [100,200,300,400,500]

df = pd.DataFrame({"Item":item,"Price":price})
print(df)

output_row = []
threshold = 250

for i in df.index:
    if df["Price"][i] > threshold:
        output_row.append(i)
        
print("Rows having value greater than threshold : ",output_row)

print(df[df["Price"]>threshold])
```

# 10.Create a histogram using Seaborn to visualize a distribution.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np


data = np.random.normal(loc=50, scale=10, size=1000)

sns.set(style="whitegrid")  
plt.figure(figsize=(8, 5))

sns.histplot(data, bins=30, kde=True, color='blue', edgecolor='black')

plt.title("Distribution of Data", fontsize=16)
plt.xlabel("Value", fontsize=12)
plt.ylabel("Frequency", fontsize=12)

plt.show()
```

# 13.Create a 3D scatter plot using Plotly.

```python
import plotly.graph_objects as go

x = [1,2,3,4,5]
y = [1,4,9,16,25]
z = [2,6,12,20,30]

fig = go.Figure(data = [go.Scatter3d(x=x,y=y,z=z,mode='markers')])

fig.update_layout(
    title="3D Scatter Plot",
    scene=dict(
        xaxis_title="X Axis",
        yaxis_title="Y Axis",
        zaxis_title="Z Axis"
    ),
    template="plotly"
)

fig.show()
```
